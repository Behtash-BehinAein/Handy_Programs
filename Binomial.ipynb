```python
import scipy.special
import numpy as np
from itertools import permutations 
import matplotlib.pyplot as plt
```

# In this notebook

# Functional to extract binomial distribution

```python
def Binomial(n, k , p):
    
    Bin_Coef        = scipy.special.binom(n, k)            # Binomial coefficient
    PMF_with_p_atK  = Bin_Coef * (p**k) * ((1-p)**(n-k))   # Probability of k-successes in n trials 
    
    return Bin_Coef , PMF_with_p_atK

```

# CALCULATOR: Specific points on pmf

```python
n , k , p  = 10, 6, 0.5
Coef, Prob = Binomial(n , k , p)

print(f'- Binomial Coefficient is : {int(Coef)}' ) 
print(f'- PMF for {k} successes in {n} trials with p={p} is  : {round(Prob,3)} ')
```

# pmf

```python
n1 , n2      = 20 , 40        ; n = [n1, n2]
p1 , p2, p3  = 0.5, 0.3 , 0.7 ; p = [p1 , p2, p3]

# Form all unique combinations of the parameters
parameters = [list(zip(n , p_permuated)) for p_permuated in permutations(p) ]
lst = list(set([ii[0] for ii in parameters] +  [ii[1] for ii in parameters]))


num_dists = len(lst)
dists = {}
fig   = plt.figure(figsize=(18,10))
ncols = 3
nrows = num_dists//ncols

for ii in range(len(lst)): 
    n , p = lst[ii][0] , lst[ii][1] 
    k = np.linspace(0 , n, n+1).astype(int)    
    dists[ii] = [n,p]  
    
   
    fig.add_subplot(nrows , ncols, ii+1)
    _ , pmf = Binomial(n, k, p)
    plt.plot( k , pmf ,'b.--', label=f'n={n} , p={p}')
    plt.axvline(x=20, color='k')
    plt.xlim((0,n2))
    plt.xlabel('# of successes', fontsize=16)
    plt.ylabel('Probability', fontsize=16)
    plt.grid()
    plt.legend(fontsize=20)
```
